# Pool Countries using two countries of different basins (SI/NA)

I chose two countries (Comores and St. Kitts & Nevis) and created for each country a cat bond. Afterwards I pooled the countries by adding the principals from the single country bond and created a multi-country cat bond. <br>
For all cat bonds I simulated the expected losses, attachment probability and coverage. Furthermore, I implemented various pricing approaches. Additionally, I simulated each bond using different principals, ranging from 1% of the countries GDP up until 20%.

## Coverage 
The following plots ilustrates the coverage (sum of payouts / sum of damages) for each country attained through either a single country bond or a pooled multi country bond. Additionally, the coverage per paid premium is showed country specific for single and multi country bonds. Ciullos risk sharing approach was used to split premiums.



Coverage across single and multi country bonds for the individual countries are very similar. Coverage for the multi country bond is most of the times a bit higher, but only marginal. This is possibly due to the fact that there is a higher principal availabale and in the unlikely event of two major events during one term more coverage can be achieved. 
Not surprisingly, the higher the principal the higher the coverage, but with increasing principal the additionally attained coverage decreases.

Using the Expected shortfall with an alpha of 0.95 seems a fair approach to split premiums as both countries can purchase more coverage with less money at least most of the time.
